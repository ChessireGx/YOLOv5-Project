```python
import torch
from ultralytics import YOLO

# 1. Cargar el modelo YOLOv8
model = YOLO(model="/content/GroundingDINO/runs/detect/train5/weights/best.pt")  # Carga el modelo YOLOv8

# 2. Definir una entrada de prueba (dummy input)
dummy_input = torch.randn(1, 10)  # Ajustar según las dimensiones de entrada de tu modelo

# 3. Exportar el modelo a formato ONNX
onnx_path = "modelo.onnx"
torch.onnx.export(
    model,                          # Modelo entrenado
    dummy_input,                    # Entrada de prueba
    onnx_path,                      # Ruta de salida
    export_params=True,             # Exportar parámetros entrenados
    opset_version=11,               # Versión de ONNX
    do_constant_folding=True,       # Optimizar constantes en el modelo
    input_names=["input"],          # Nombre de la entrada
    output_names=["output"],        # Nombre de la salida
    dynamic_axes={                  # Definir ejes dinámicos
        "input": {0: "batch_size"},
        "output": {0: "batch_size"}
    }
)

print(f"Modelo exportado a ONNX en: {onnx_path}")
```
